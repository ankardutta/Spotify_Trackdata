# Q-2 TOP 10 MOST POPULAR SONGS GREATER THAN 90 VALUE

This problem can be solved 

In two ways -- 1. Using FILTER OPTIONS
               2. using QUERY() method

```python
# using FILTER OPTIONS

df_spot[df_spot.popularity >90].head(10)
```

**The query() method allows you to query the DataFrame.**

**The query() method takes a query expression as a string parameter, which has to evaluate to either True of False.**

It returns the DataFrame where the result is True according to the query expression.

```python
# using QUERY() METHOD

df_spot.query('popularity>90').sort_values("popularity", ascending=False).head(10)
```
